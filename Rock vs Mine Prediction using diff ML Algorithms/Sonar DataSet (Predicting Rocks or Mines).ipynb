# Model Evaluation

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

# train_test split ways

```python
arr = data.values
X = arr[:,0:-1]
y = arr[:,-1]

testsize = 0.2
print(y)
```

```python
X = data.drop('R',axis=1)
y = data['R']
print(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=testsize)
```

```python
print(y_test)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
acc = accuracy_score(y_test,y_pred)*100
print(f'Decision Tree Accuracy Score is: {acc}')
```

```python
rt = RandomForestClassifier()
rt.fit(X_train,y_train)
y_preds = rt.predict(X_test)
accu = accuracy_score(y_test,y_preds)*100
print(f'Random Forest Accuracy Score is: {accu}')
```

```python
models = []
models.append(('DT',DecisionTreeClassifier()))
models.append(('RF',RandomForestClassifier()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('LR',LogisticRegression()))
models.append(('NB',GaussianNB()))
```

```python
print(models)
```

```python
names = []
results = []
```

```python
for name,model in models:
    obj = model
    names.append(name)
    obj.fit(X_train,y_train)
    y_predict = obj.predict(X_test)
    results.append(accuracy_score(y_test,y_predict)*100)
    print(results)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(names,results)
```

# DEEP LEARNING

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
data['R'].replace(to_replace=['R','M'],value=[0,1],inplace=True)
X = data.drop('R',axis=1)
y = data['R']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=testsize)
```

```python
ann_model = Sequential([
    Dense(24,input_dim=60, activation='relu'),
    Dense(12,activation='relu'),
    Dense(8,activation='relu'),
    Dense(1,activation='sigmoid') # last layer sigmoid is used
])

ann_model.compile(
    optimizer = keras.optimizers.Adam(),
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
ann_model.summary()
```

```python
ann_model.fit(X_train,y_train,epochs=20)
```

```python
loss, acc = ann_model.evaluate(X_test,y_test)
```

```python
print(loss)
print(acc)
names.append('ANN')
```

```python
results.append(acc*100)
```

```python
print(results)
```

```python
plt.scatter(names, results)
```
